```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    'Sleep': ['Good','Good','Good','Good','Poor','Poor','Poor','Poor'],
    'Practice': ['Yes','Yes','No','No','Yes','Yes','No','No'],
    'Revision': ['Yes','No','Yes','No','Yes','No','Yes','No'],
    'Result': ['Pass','Pass','Pass','Fail','Fail','Fail','Fail','Fail']
}

df = pd.DataFrame(data)

# Step 2: Encode categorical values
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])

# Step 3: Split features and target
X = df[['Sleep', 'Practice', 'Revision']]
y = df['Result']

# Step 4: Train Decision Tree using Entropy (ID3)
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X, y)

# Step 5: Visualize the Decision Tree
plt.figure(figsize=(12,8))
tree.plot_tree(
    model,
    feature_names=['Sleep', 'Practice', 'Revision'],
    class_names=['Fail', 'Pass'],
    filled=True
)
plt.show()

# Step 6: Predict for a new instance
# Example: Sleep = Good, Practice = No, Revision = Yes
new_data = pd.DataFrame({
    'Sleep': [1],     # Good
    'Practice': [0],  # No
    'Revision': [1]   # Yes
})

prediction = model.predict(new_data)

if prediction[0] == 1:
    print("Prediction: Pass")
else:
    print("Prediction: Fail")
```
